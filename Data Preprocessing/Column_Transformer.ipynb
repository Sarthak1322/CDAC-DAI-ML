```python
df = pd.read_csv(r"/home/sarthakredasani/Documents/CDAC_ML/Datasets/Datasets/Housing.csv")
df.head(5)
dum_df = pd.get_dummies(df, drop_first=True)
```

```python
ohe = OneHotEncoder(sparse_output=False, drop='first').set_output(transform='pandas')
df.dtypes
```

```python
str_cols = ['driveway', 'recroom' , 'fullbase', 'gashw', 'airco', 'prefarea']
num_cols = ['price', 'lotsize', 'bedrooms', 'bathrms', 'stories', 'garagepl']
ct = make_column_transformer((ohe, str_cols), ("passthrough", num_cols), verbose_feature_names_out=False)
ct.set_output(transform='pandas')
```

```python
ohe_df = ct.fit_transform(df)
ohe_df
```

```python
# Using make_column_selector
```

```python
ct = make_column_transformer((ohe, make_column_selector(dtype_include=object)), ("passthrough", make_column_selector(dtype_exclude=object)), verbose_feature_names_out=False)
ct = ct.set_output(transform='pandas')
ohe_df = ct.fit_transform(df)
ohe_df
```

```python
lr = LinearRegression()
pipe = Pipeline([('TRANS', ct), ('LR', lr)])
X, y = df.drop('price', axis=1), df['price']
```

```python
pipe.fit(X,y)
```

```python
import pandas as pd 
import numpy as np
from sklearn.linear_model import *
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
import os
```

```python
df = pd.read_csv(r"/home/sarthakredasani/Documents/CDAC_ML/Datasets/Datasets/Exp_Salaries.csv")
df.head()
```

```python
df_dum = pd.get_dummies(df, drop_first=True)
X, y = df_dum.drop('Salary', axis =1), df_dum['Salary']
```

```python
df_dum
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=25)
```

```python
X_train.shape , y_train.shape
```

```python
# let's do ridge

ridge = Ridge(alpha=0.01)
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_test)
print(r2_score(y_test, y_pred))   
```

```python
# let's try for other alpha values in for loop
alphas = [0.001, 0.01, 0.1, 0.5, 1, 2]
scores = []
for i in alphas:
    ridge = Ridge(alpha=i)
    ridge.fit(X_train, y_train)
    y_pred = ridge.predict(X_test)
    scores.append(r2_score(y_test, y_pred))
```

```python
scores
```

```python
#  using lasso
score = []
for i in alphas:
    lasso = Lasso(alpha=i)
    lasso.fit(X_train, y_train)
    y_pred = lasso.predict(X_test)
    score.append(r2_score(y_test, y_pred))
```

```python
score
```

```python
# let's use ElasticNet(L1 + L2)

ratios = [0.01, 0.25, 0.5, 0.75, 0.9]
alphas = [0.001, 0.01, 0.1, 0.5, 1, 2]

scoreEn = []
for a in alphas:
    for r in ratios:
        el = ElasticNet(alpha=a, l1_ratio = r)
        el.fit(X_train, y_train)
        y_predEn = el.predict(X_test)
        scoreEn.append([r, a, r2_score(y_test, y_predEn)])
```

```python
scoreEn
```

```python
df_scoreEn = pd.DataFrame(scoreEn, columns=['ratio', 'alpha', 'score'])
df_scoreEn.sort_values('score', ascending=False).head(1)
```

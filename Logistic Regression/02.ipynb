```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
from sklearn.linear_model import *
from sklearn.metrics import *
from sklearn.model_selection import *
import matplotlib.pyplot as plt
```

```python
sonar = pd.read_csv(r"/home/sarthakredasani/Documents/CDAC_ML/Datasets/Datasets/Sonar.csv")
sonar.head(5)
```

```python
# Target column is Class
```

```python
X, y = sonar.drop('Class', axis =1), sonar['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
lr = LogisticRegression(penalty='l2', solver='lbfgs')
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
lr = LogisticRegression(penalty=None, solver='lbfgs')
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
# let's run for loop

penalties = ['l1', 'l2', 'elasticnet', None]
ratios = np.linspace(0.001,0.98, 20)

scores =[]

for i in penalties:
    for j in ratios:
        lr = LogisticRegression(penalty=i, solver='saga', l1_ratio=j)
        lr.fit(X_train, y_train)
        y_pred = lr.predict(X_test)
        scores.append([i,j,(accuracy_score(y_test, y_pred))])
df_scores = pd.DataFrame(scores, columns=['Penality', 'Ratios', 'score'])
df_scores.sort_values('score', ascending=False).head(5)
```

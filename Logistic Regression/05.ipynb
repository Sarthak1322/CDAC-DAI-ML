```python
from sklearn.linear_model import *
from sklearn.metrics import *
from sklearn.preprocessing import *
from sklearn.model_selection import *
from sklearn.pipeline import *
from sklearn.compose import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = pd.read_csv(r"/home/sarthakredasani/Documents/CDAC_ML/Cases/Cases/Image Segmentation/Image_Segmentation.csv")
```

```python
image.head(5)
```

```python
X, y = image.drop('Class', axis =1), image['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
y_test.value_counts()
```

```python
lr = LogisticRegression(penalty='l2')
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
lr.coef_ , lr.intercept_
```

```python
penaliies = ['l1', 'l2', 'elasticnet', None]
ratios = np.linspace(0.001, 0.98, 20)
scores = []
for i in penaliies:
    for j in ratios:
        lr = LogisticRegression(penalty=i, l1_ratio=j, solver='saga')
        lr.fit(X_train, y_train)
        y_pred = lr.predict(X_test)
        scores.append([i , j, (accuracy_score(y_test, y_pred))])
df_scores = pd.DataFrame(scores, columns=['Penality', 'Ratio', 'Score'])
df_scores.sort_values('Score', ascending=False).head(5)
```

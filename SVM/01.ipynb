```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import recall_score, precision_score, f1_score
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
sonar = pd.read_csv(r"/home/sarthakredasani/Documents/CDAC_ML/Cases/Cases/Sonar/Sonar.csv")
sonar.head(5)
```

```python
X, y = sonar.drop('Class', axis=1), sonar['Class']
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25, stratify=y)
```

```python
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
Cs = np.linspace(0.01, 5, 20)
scores = []
for c in Cs:
    svm = SVC(kernel='linear',C=c)
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    scores.append([c,(accuracy_score(y_test, y_pred))])
```

```python
df_scores = pd.DataFrame(scores, columns=['C', 'score'])
df_scores.sort_values('score', ascending=False)
```

```python
# Radial Kernel
```

```python
svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
Cs = np.linspace(0.01, 5, 20)
Gs = np.linspace(0.01, 5, 20)
scores = []
for c in Cs:
    for g in Gs:
        svm = SVC(kernel='rbf', C=c, gamma=g)
        svm.fit(X_train, y_train)
        y_pred = svm.predict(X_test)
        scores.append([c,g,(accuracy_score(y_test, y_pred))])
```

```python
df_scores = pd.DataFrame(scores, columns=['C','Gamma', 'score'])
df_scores.sort_values('score', ascending=False)
```
